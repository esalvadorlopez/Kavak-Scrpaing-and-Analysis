```python
import pandas as pd
import json
import gspread
```

```python
df = pd.read_csv('raw_dataset.csv')
df.sample(5)
```

```python
df = df.iloc[:,1:]
df.sample(5)
```

```python
df['features'] = df['features'].str.replace("'",'"')
df['features'] = df['features'].str.replace(" ",'')
df['features'] = df['features'].apply(lambda x: json.loads(x))
```

```python
count=0
for row in df['features']:
    row.update({'key': count})
    count +=1
df['features']
```

```python
features_list = []
for row in df['features']:
    features_list.append(row)
df2 = pd.DataFrame(features_list)
df2
```

```python
df['key'] = df.index
df.head()
```

```python
df_final = df.merge(df2, on='key',how='left')
```

```python
df_final.sample(5)
```

```python
df_final.drop(columns = ['key','features'], inplace=True)
```

```python
df_final.sample(5)
```

```python
df_final['title'] = df_final['title'].str.replace(r' #[0-9]+.*$','',regex=True)
df_final['modelo'] = df_final['title'].str.extract(r'(\s\d+$)')
df_final['title'] = df_final['title'].str.replace(r'\s\d+$','',regex=True)
df_final['brand'] = df_final['title'].str.extract(r'(^[A-Z-a-z]+) .*')
```

```python
df_final['km'] = df_final['km'].str.replace(r'\D','',regex=True)
df_final['price'] = df_final['price'].str.replace(r'\D','',regex=True)
# if 'Litros' in df_final:
#     df_final['Litros'] = df_final['Litros'].str.replace(',','.')
# if 'ConsumodeGasolina' in df_final:
#     df_final['ConsumodeGasolina'] = df_final['ConsumodeGasolina'].str.replace(',','.')
# if 'NumerodeVelocidades' in df_final:
#     df_final['NumerodeVelocidades'] = df_final['NumerodeVelocidades'].str.replace(r'\D','',regex=True)
```

```python
df_final.sample(5)
```

```python
df_final = df_final[['StockID','title','brand','modelo','Transmisión','km','price','url','image_link']]
```

```python
df_final.sample(5)
```

```python
df_final = df_final.fillna('')
df_final = df_final.drop_duplicates()
```

```python
df_final = df_final.rename(columns={
    'StockID' : 'Stock Id',
    'title' : 'Automovil',
    'brand' : 'Marca',
    'modelo' : 'Modelo (Año)',
    'km' : 'Kilometraje',
    'price' : 'Precio',
    'url' : 'URL',
    'image_link' : 'URL de la Imagen'

})
```
